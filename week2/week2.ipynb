```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
def sharpen_image(original_img, kernel_size=(7, 7), gamma=2):
    # TODO: Apply a smoothing filter (e.g., Gaussian blur)
    #blur = cv2.GaussianBlur(image, kernel_size, 0)
    blur = cv2.GaussianBlur(original_img, (7, 7), 0)

    # TODO: Calculate the difference between the original and blurred image
    detail = cv2.subtract(original_img, blur)

    # TODO: Apply the gamma factor
    sharpened_detail = cv2.multiply(detail, np.array([gamma]))

    # TODO:
    # Add the sharpened detail back to the original image
    # Convert sharpened_detail to the same data type as the original image
    #sharpened_img = cv2.add(image, sharpened_detail)
    sharpened_img = cv2.addWeighted(original_img, 1 + gamma, blur, -gamma, 0)


    # 5. Ensure correct data type (uint8 in range 0-255)
    sharpened_img = np.clip(sharpened_img, 0, 255).astype(np.uint8)

    return sharpened_img, blur


image = cv2.imread('tiger.jpg')
if image is not None:
  sharpened_img, blur = sharpen_image(image)
  cv2_imshow(image)
  cv2_imshow(sharpened_img)
  cv2_imshow(blur)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
else:
  print("Error: Could not load image.")

#-------------------------------------
# def sharpen_image(image, gamma=1):
#     # TODO: Apply a smoothing filter (e.g., Gaussian blur)

#     # TODO: Calculate the difference between the original and blurred image
#     kernel = np.array([[0,       -gamma,       0],
#                        [-gamma,  1 + 4*gamma, -gamma],
#                        [0,       -gamma,       0]])

#     # TODO: Apply the gamma factor

#     # TODO: Add the sharpened detail back to the original image
#     sharpened_img = cv2.filter2D(image, -1, kernel)

#     # TODO: Convert sharpened_detail to the same data type as the original image
#     sharpened_img = np.clip(sharpened_img, 0, 255).astype(np.uint8)

#     return sharpened_img
```

```python
# Read image
img = cv2.imread("bacteria.jpg", cv2.IMREAD_GRAYSCALE)

# Binarize
_, binary = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Kernels
kernel = np.ones((5, 5), np.uint8)
kernel2 = np.ones((2 ,2), np.uint8)

# Morphological edge detection (erosion)
eroded_img = cv2.erode(binary, kernel, iterations=1)
morph_edge = cv2.subtract(binary, eroded_img)

# Clean edges with small erosion
clean_edges = cv2.erode(morph_edge, kernel2, iterations=1)

# Show results
cv2_imshow(binary)
cv2_imshow(morph_edge)
cv2_imshow(clean_edges)
```

```python
# Load image (grayscale)
img = cv2.imread("bacteria.jpg", cv2.IMREAD_GRAYSCALE)

# Otsu thresholding
_, binary = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# Flip the colors (so bacteria are white)
binary = cv2.bitwise_not(binary)

# Erosion (17 iterations as you tested)
kernel = np.ones((3,3), np.uint8)
cleaned = cv2.erode(binary, kernel, iterations=17)

# Connected Components
num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(cleaned, connectivity=8)
print("Number of components (excluding background):", num_labels - 1)

output_img = cv2.cvtColor(cleaned, cv2.COLOR_GRAY2BGR)
for i in range(1, num_labels):  # skip background
    x, y, w, h, area = stats[i]
    if area > 10:  # filter very small blobs
        cv2.rectangle(output_img, (x, y), (x+w, y+h), (0, 0, 255), 1)
        # Draw centroid as a small green circle
        cx, cy = int(centroids[i][0]), int(centroids[i][1])
        cv2.circle(output_img, (cx, cy), 3, (0, 255, 0), -1)  # radius=3, filled


# Display results
cv2_imshow(img)        # Original image
cv2_imshow(binary)     # Binary inverted
cv2_imshow(cleaned)    # After erosion (17 iterations)
cv2_imshow(output_img) # Final result with bounding boxes
```
